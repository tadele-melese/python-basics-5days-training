### A. üìù <span style='color:blue'>Comments </span>

- Comments are notes written in the code for humans, not for the computer.
- Python ignores comments when running the program.
- Used to explain code, make it readable, or disable lines temporarily.

#### 1. Single-Line Comment
- Begins with #.
- Python ignores everything after # on that line.

```python
# The value 10 is stored in X variable
x = 10  
x
```

#### 2. Multi-Line Comment
- Python has no official multi-line comment syntax.
- We use triple quotes """ ... """ or ''' ... ''' to write comments across many lines.
- These are actually unused multiline strings but function as comments.

```python
"""
This is a multi-line comment.
You can write many lines here.
Useful for long explanations.
"""
```

#### 3 3. Inline Comment
- A comment written at the end of a line of code using #.

```python
speed = 80  # speed in km/h
```

### C üìè  <span style='color:blue'>Python Indentetion </span>

**What is Indentation?**
- Indentation means adding spaces (or a tab) at the beginning of a line.
- In Python, indentation is NOT optional. It is part of the syntax.
- It tells Python which lines of code belong together.
- Instead, **Python** uses indentation to define blocks of code, such as:
     - **loops**
     - **<span style="color:green">if </span>statements**
     - **functions**
     - **classes**
- If indentation is incorrect ‚Üí Python will show an error.
‚úî **Default Rule**
- Use 4 spaces for indentation.
- Never mix spaces and tabs.


üß© **Example 2: Indentation in a Loop
for i in range(5):
    print(i)        # inside loop
print("Done")        # outside loop

üß© Example 3: Indentation in a Function
def greet(name):
    print("Hello,", name)

```python
# Example 1: Indentation in an If Statement
# Correct:
if age > 18:
    print("Adult")
```

```python
# Example 2: Indentation in a Loop
for i in range(5):
    print(i)
```

```python
# Incorrect (no indentation):
if age > 18:
print("Adult")   # ‚ùå üß©Error
```

## **Comparison Operators**

- Used to compare values (result is True/False).

| Operator | Meaning          | Example         |
| -------- | ---------------- | --------------- |
| `==`     | Equal            | `5 == 5` ‚Üí True |
| `!=`     | Not equal        | `5 != 3` ‚Üí True |
| `>`      | Greater than     | `5 > 3` ‚Üí True  |
| `<`      | Less than        | `5 < 3` ‚Üí False |
| `>=`     | Greater or equal | `6 >= 6` ‚Üí True |
| `<=`     | Less or equal    | `3 <= 5` ‚Üí True |

```python
# Examples
x = 5
y = 3

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

## **Logical Operators**

- Used to combine conditions.

| Operator | Meaning   | Example          | Result |
| -------- | --------- | ---------------- | ------ |
| `and`    | Both True | `True and False` | False  |
| `or`     | One True  | `True or False`  | True   |
| `not`    | Reverse   | `not True`       | False  |

```python
a = True
b = False
print(a and b)
print(a or b)
print(not a)
```

## **Membership Operators**

- Used to test if a value is in a sequence.

| Operator | Meaning        | Example                   |
| -------- | -------------- | ------------------------- |
| `in`     | Exists in      | `"a" in "cat"` ‚Üí True     |
| `not in` | Does not exist | `5 not in [1,2,3]` ‚Üí True |

```python
list1 = [1, 2, 3, 4]
print(2 in list1)
print(5 in list1)
```

## **Identity Operators**

- Used to compare memory locations.

| Operator | Meaning         | Example      |
| -------- | --------------- | ------------ |
| `is`     | Same object     | `x is y`     |
| `is not` | Not same object | `x is not y` |

```python
# Examples
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)
print(a is c)
```

‚úî Solution
- a is b ‚Üí True (same memory location)
- a is c ‚Üí False (same values, different memory)
