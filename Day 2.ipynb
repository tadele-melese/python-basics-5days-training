##  **<span style="color:blue">2.1. Lists </span>**  

- What is a **List**?
   - A list is a collection of items stored in a single variable.
   - It is ordered, changeable (mutable), and allows duplicate items

```python
my_list = [10, "apple", 3.14, True]
```

```python
print(my_list)
```

### **Key Features of Lists**

| Feature           | Meaning                                        |
| ----------------- | ---------------------------------------------- |
| **Ordered**       | Items have fixed positions (index starts at 0) |
| **Mutable**       | You can change, add, remove items              |
| **Heterogeneous** | Items can be of different data types           |
| **Dynamic**       | You can grow/shrink the list anytime           |

##  **<span style="color:blue">2.2. Tuples</span>**

- **Tuple** items are indexed, the first item has index [0], the second item has index [1] etc.

##  **<span style="color:blue">2.3. Dictionaries</span>**  

**What is a Dictionary?**
 - A **dictionary** in Python is a data structure that stores data in **keyâ€“value** pairs.
 - **Key** = the name (must be unique)
 - **Value** = the data related to that key
 -  A dictionary is a collection which is **ordered**, **changeable** and **do not allow duplicates**.

```python
# Example
student = {
    "name": "Biniam",
    "age": 25,
    "country": "Ethiopia"
}
```

**Why Dictionaries Are Useful?**
 - They help you store structured data.
 - Fast for searching, updating, and organizing information.
 - Used in **machine learning**, **GIS**, **finance**, **web development**, **APIs**, etc.

**Creating a Dictionary**

```python
car = {
    "brand": "Toyota",
    "model": "Corolla",
    "year": 2020
}

# or 

car = dict(brand="Toyota", model="Corolla", year=2020)
```

**Accessing Dictionary Values**

```python
print(car["brand"])
```

```python
print(car.get("model"))  
```

**Adding & Updating Values**

```python
car["color"] = "White"           # Add new key-value
```

```python
car["year"] = 2023               # Update value
```

**Removing Items**

```python
car.pop("model")        # Remove specific key
```

```python
car.popitem()           # Removes last inserted item (Python 3.7+)
```

```python
del car["brand"]        # Delete key
```

```python
car.clear()             # Remove all items
```

**Looping (Iterating) Through Dictionaries**

```python
# loop through keys
for key in car:
    print(key)
```

```python
# loop through values
for value in car.values():
    print(value)
```

```python
# Loop through keyâ€“value pairs:
for key, value in car.items():
    print(key, value)
```

**Checking if a Key Exists**

```python
if "brand" in car:
    print("Brand exists!")
```

**Nested Dictionaries (Dictionary inside a Dictionary)**

```python
students = {
    "S1": {"name": "Amanuel", "score": 90},
    "S2": {"name": "Sara", "score": 85}
}
```

```python
# Accessing
print(students["S1"]["name"])      # Output: Amanuel
```
