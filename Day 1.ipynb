```python
# Interactive Smart Irrigation Advisor
# Beginner-friendly but professional-level logic

import time

print("ðŸŒ± Smart Irrigation Advisor v2.0")
print("ðŸ¤– Welcome! I will help you decide when and how to irrigate your crops.\n")

# User inputs
crop = input("ðŸŒ¾ Enter crop type (e.g., maize, wheat, teff): ").strip().lower()
soil_type = input("ðŸª¨ Enter soil type (sand / loam / clay): ").strip().lower()
moisture_level = int(input("ðŸ’§ Enter current soil moisture level (0â€“100%): "))
temperature = float(input("ðŸŒ¡ï¸ Enter today's temperature (Â°C): "))

time.sleep(1)

print("\nðŸ“¡ Analyzing field conditions...")
time.sleep(1)

# Soil water holding capacity
soil_factor = {
    "sand": 1.5,
    "loam": 1.0,
    "clay": 0.7
}

# Crop water sensitivity
crop_factor = {
    "maize": 1.3,
    "wheat": 1.1,
    "teff": 1.0
}

sf = soil_factor.get(soil_type, 1.0)
cf = crop_factor.get(crop, 1.0)

# Decision logic
stress_index = (100 - moisture_level) * sf * cf

if stress_index > 80 or moisture_level < 30:
    decision = "IRRIGATION REQUIRED"
    water_liters = round(stress_index * (1 + temperature / 40), 1)

elif 40 <= stress_index <= 80:
    decision = "MONITOR CLOSELY"
    water_liters = 0

else:
    decision = "NO IRRIGATION NEEDED"
    water_liters = 0

# Intelligent output
print("\nðŸ§  SYSTEM DECISION")
print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
print(f"ðŸŒ¾ Crop: {crop.capitalize()}")
print(f"ðŸª¨ Soil Type: {soil_type.capitalize()}")
print(f"ðŸ’§ Moisture Level: {moisture_level}%")
print(f"ðŸŒ¡ï¸ Temperature: {temperature} Â°C")
print(f"ðŸ“Š Stress Index: {round(stress_index,1)}")

print(f"\nâš™ï¸ Recommendation: {decision}")

if water_liters > 0:
    print(f"ðŸ’¦ Suggested Irrigation Volume: {water_liters} liters")
    print("ðŸš¿ Action: Activate irrigation system")
else:
    print("âœ… Action: No irrigation required at this time")

print("\nðŸ“ Decision stored for farm management records.")
print("ðŸ¤– Advisor session completed.")
```

**What is Indentation?**
- Indentation means adding spaces (or a tab) at the beginning of a line.
- In Python, indentation is NOT optional. It is part of the syntax.
- It tells Python which lines of code belong together.
- Instead, **Python** uses indentation to define blocks of code, such as:
     - **loops**
     - **<span style="color:green">if </span>statements**
     - **functions**
     - **classes**
- If indentation is incorrect â†’ Python will show an error.
âœ” **Default Rule**
- Use 4 spaces for indentation.
- Never mix spaces and tabs.
