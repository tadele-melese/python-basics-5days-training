## **<span style="color:blue">3.1 Conditional Statements</span>**  

**Conditional statements** allow you to make decisions in code based on conditions (True/False).

###  **<span style="color:blue">3.1.1 The if Statements</span>**  

**Syntax**

```python
if condition:
    # code block
```

```python
# Example
temperature = 38
if temperature > 35:
    print("It is very hot today.")
```

###  **<span style="color:blue">3.1.2 The ifâ€“else Statement</span>**  

**Syntax**

```python
if condition:
    # do A
else:
    # do B
```

```python
# Example1
ndvi = 0.15
if ndvi > 0.3:
    print("Good vegetation condition")
else:
    print("Poor vegetation condition")
```

```python
# Example2
age = input("Your age: ")
if age >= 18:
    print("Hence, you are eligible to vote")
else:
    print("Not eligible")
```

###  **<span style="color:blue">3.1.3 The elif Statement</span>**  

- Used when you have **multiple conditions** to check.

**Syntax**

```python
if condition1:
    ...
elif condition2:
    ...
elif condition3:
    ...
else:
    ...
```

```python
# Example 1
score = 67
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D")
```

```python
# Example 2
river_level = 4.2  # meters

if river_level < 2:
    print("Low flow")
elif river_level < 4:
    print("Normal flow")
elif river_level < 6:
    print("High flow")
else:
    print("Flooding risk!")
```

```python
# Example 3
soil_moisture = 18

if soil_moisture < 15:
    print("Severe Dryness")
elif soil_moisture < 25:
    print("Dry")
elif soil_moisture < 40:
    print("Optimal")
else:
    print("Wet")
```

###  **<span style="color:blue">3.1.4 Nested Conditions</span>**  

- **Condition inside another condition**

**Syntax**

```python
if condition:
    if another_condition:
        ...
    else:
        ...
else:
    ...
```

```python
# Example 1
age = 20
citizen = 'Ethiopia'

if age >= 18:
    if citizen:
        print("Eligible to vote")
    else:
        print("Age OK but not a citizen")
else:
    print("Not eligible")
```

```python
# Example 2
cow_temp = 39.8
feed_intake = "low"

if cow_temp > 39.5:
    if feed_intake == "low":
        print("High fever and reduced appetite â€“ immediate treatment needed!")
    else:
        print("Fever detected â€“ monitor closely.")
else:
    print("Normal health.")
```

## **<span style="color:blue">3.2 Loops</span>**  

- **Loops** allow us to repeat a block of code multiple times without writing it repeatedly.

### **<span style="color:blue">3.2.1. For Loops</span>**  

- The **for** loop iterates over a sequence (list, tuple, set, dict, string).

**Syntax**

```python
for variable in sequence:
    # code block
```

```python
# Example 1
fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    print(fruit)
```

```python
# Example2
climate_vars = ["temperature", "rainfall", "humidity", "wind_speed"]

for var in climate_vars:
    print("Monitoring:", var)
```

```python
# Example 3
soil_ph = [5.2, 6.1, 4.8, 7.0, 5.5]

for ph in soil_ph:
    if ph < 5.5:
        print(f"Soil pH {ph}: Acidic â†’ Add lime")
    elif ph > 7:
        print(f"Soil pH {ph}: Alkaline â†’ Add sulfur")
    else:
        print(f"Soil pH {ph}: Optimal")
```

### **<span style="color:blue">3.2.2. While Loops</span>** 

- **while** loops run as long as a condition is **True**.

**Syntax**

```python
while condition:
    # code block
```

```python
# Example1
count = 1
while count <= 5:
    print("Count:", count)
    count += 1   # count = count + 1
```

```python
# Example 2
river_level = 5.5
rain_forecast = True

if river_level > 5:
    if rain_forecast:
        print("Severe Flood Risk! Evacuate.")
    else:
        print("Flood Risk â€“ monitor conditions.")
else:
    print("Normal conditions.")
```

```python
# Example 3
river_level = 3.5

while river_level < 5:
    print(f"River level: {river_level} m â†’ Safe")
    river_level += 0.5

print("Warning: River level critical!")
```

### **<span style="color:blue">3.2.3. Loop Control: break and continue </span>** 

**a) break**
- Stops the loop immediately, even if the condition is True.
- Example: Stop monitoring when extreme temperature detected

```python
temps = [36, 37, 38, 41, 39]

for t in temps:
    if t > 40:
        print(f"Extreme temperature {t}Â°C! Stop monitoring!")
        break
    print(f"Temperature {t}Â°C: Normal")
```

**b) continue**
-  Skips the current iteration and continues with the next one.
-  Example: Skip healthy animals in a health check

```python
cow_temps = [38.2, 39.6, 38.5, 40.0]

for temp in cow_temps:
    if temp <= 39.0:
        continue  # skip healthy cows
    print(f"Alert: Cow temperature {temp}Â°C â†’ Treat immediately")
```

### **<span style="color:blue">3.2.4. Nested Loops</span>**  

-  Loops inside loops can handle multiple dimensions of data.
-  GIS/Climate Example: Monitoring stations Ã— variables

```python
stations = ["Station_A", "Station_B"]
variables = ["temperature", "rainfall", "humidity"]

for station in stations:
    for var in variables:
        print(f"{station}: Monitoring {var}")
```

### Combined Example: Loops with Conditional + break/continue

```python
river_levels = [3.2, 3.8, 4.5, 5.2, 4.8]

for level in river_levels:
    if level > 5:
        print(f"Flood alert! Level: {level} m â†’ Evacuate!")
        break
    elif level > 4:
        print(f"High water level: {level} m â†’ Monitor closely")
        continue
    print(f"Normal level: {level} m")
```

## **<span style="color:blue">3.3 Practical examples</span>**  

**Scenario 1:** Automatically water plots until soil moisture reaches 40%.

```python
plots = {"Plot_1": 25, "Plot_2": 18, "Plot_3": 35}
```

```python
for plot, moisture in plots.items():
    while moisture < 40:
        print(f"{plot}: Watering crops (Current moisture: {moisture}%)")
        moisture += 5  # simulate irrigation
    print(f"{plot}: Optimal moisture reached ({moisture}%)\n")
```

**Scenario 2**: You are monitoring multiple rivers. 
-  Stop monitoring once a critical flood level is reached (break). Skip rivers that are normal (continue).

```python
rivers = {
    "Blue_River": 3.2,
    "Red_River": 4.5,
    "Green_River": 5.3,
    "Yellow_River": 2.8
}

for river, level in rivers.items():
    if level < 4:
        continue  # safe rivers, skip
    elif level > 5:
        print(f"{river}: Flood warning! Evacuate immediately!")
        break
    else:
        print(f"{river}: High water level â€“ monitor closely")
```

**Scenario 3**: Identify households at economic risk based on income and access to irrigation.

```python
households = [
    {"name": "HH1", "income": 1200, "irrigation": False},
    {"name": "HH2", "income": 3500, "irrigation": True},
    {"name": "HH3", "income": 1800, "irrigation": False},
    {"name": "HH4", "income": 4000, "irrigation": True},
]

for hh in households:
    if hh["income"] < 2000 and not hh["irrigation"]:
        print(f"{hh['name']}: High livelihood risk!")
    elif hh["income"] < 2000 and hh["irrigation"]:
        print(f"{hh['name']}: Medium risk, irrigation helps")
    else:
        print(f"{hh['name']}: Low risk")
```
